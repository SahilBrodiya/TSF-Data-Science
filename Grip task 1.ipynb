```python
# Importing libraries.
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
# importing data and viewing it.
df = pd.read_csv('http://bit.ly/w-data')
df
```

```python
# reviewing statistically
df.describe()
```

```python
# storing values in a list
hrs = list(df.Hours)
sc = list(df.Scores)
```

```python
# Visualising relation between Scores and number of hours
plt.scatter(hrs,sc)
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.show()
```

Task : Create the Decision Tree classifier and visualize it graphically.

```python
#importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Creating a dataframe by importing Iris dataset
df= pd.read_csv('Iris.csv')
df.head(20)
```

```python
# Checking count of null values
df.isnull().sum()
```

```python
# Since 'Id' is not going to have effect on our model, so this column is deleted.
df.drop('Id',axis = 1, inplace = True)
```

```python
df.head()
```

```python
df.describe()
```

```python
# Columns in the dataframe
df.columns
```

```python
X = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
Y = df[['Species']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=4)
```

```python
# Importing required libraries for Decision tree model and visualization
from sklearn.tree import DecisionTreeClassifier
from six import StringIO
from IPython.display import Image  
import matplotlib.image as mpimg
from sklearn import tree
import graphviz 
import pydotplus
from sklearn import metrics
```

```python
t = DecisionTreeClassifier(criterion="entropy", max_depth = 6)
t = t.fit(X_train,y_train)
```

```python
pred = t.predict(X_test)
```

```python
print("The accuracy of Decision Tree is:", metrics.accuracy_score(pred,y_test) * 100 ,'%')
```

```python
feature_cols = X.columns
target_cols = list(t.classes_)
```

```python
#  DECISION TREE
plt.figure(figsize = (15,8))
tree.plot_tree(t.fit(X,Y), feature_names= feature_cols , class_names = target_cols, filled = True, precision = 3,proportion =
               True, rounded = True)
plt.show()
```
